# Coin Flipping Task

### Import Packages

```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
```

### Task 1 - Flip 20 coins 10 times

```python
# Set up variables
number_of_coins = 20 #represent number of events in each simulation
simulations = 10 #represent our sample size
p = 0.5         #depicts probaility hat each coin flip will be heads(50%)


# Simulate the number of heads from flipping 20 coins 10 times
simulation_results = np.random.binomial(n=number_of_coins, p=p, size=simulations)
simulation_results #show how many heads we get from flipping 20 coins in 10 simulations
```

```python
# Test whether each simulation generated exactly 10 heads
simulation_has_10_heads = np.equal(simulation_results, 10)
simulation_has_10_heads
```

```python
# Calculate the observed probability of getting 10 heads in any given simulation
ten_heads_probability = np.equal(simulation_results, 10).mean()
#Given boolean values True and Falsoe represnet 1 and 0 respectively, we use this in computing the mean of our True-False array
ten_heads_probability
```

```python
# Calculate the observed probabilities of each number of heads from 0 to 20
probabilities = [np.equal(simulation_results, i).mean() for i in range(number_of_coins)]
probabilities
```

```python
# Plot the observed probability distribution
plt.plot(list(range(number_of_coins)),probabilities, color='orange')
plt.xlabel('Number of Heads', fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.show()
```

Probability distribution is random due to the small sample size. The sample size not confidently simulate the probaility distribution.

### Task 2 - Flip 20 coins 10,000 times

```python
# Set up variables
number_of_coins = 20 #represent number of events in each simulation
simulations = 10000 #represent our sample size
p = 0.5         #depicts probaility hat each coin flip will be heads(50%)



# Simulate the number of heads from flipping 20 coins 10,000 times

simulation_results = np.random.binomial(n=number_of_coins, p=p, size=simulations)
simulation_results 

# Calculate the probabilities of each simulation result
probabilities = [np.equal(simulation_results,i).mean() for i in range(number_of_coins)]
probabilities

# Plot the probability distribution
plt.plot(list(range(number_of_coins)),probabilities, color='orange')
plt.xlabel('Number of Heads', fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.show()

```

Probability distribution is close to a normal distribution. Increasing sample size from 10 simulations to 10,000 simulations increases the confidence of our predictions due to increae in randomness of our simulations.

By repeating a randomized process multiple times,observed probabilities start to resmeble accurate distribution.
